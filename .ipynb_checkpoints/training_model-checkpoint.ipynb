```python
 # list of all dataset in the particular order
classes = ["angry","disgust","fear","happy","neutral","sad","surprise"]
```

```python
# Read all images
for category in classes:
    #get the dir as per category
    path = os.path.join(Datadirectory,category)
    #get img in dir path
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img))
        #backtorgb = cv2.cvtColor(img_array,cv2.COLOR_GRAY2BGR)
        #plot imgs
        plt.imshow(cv2.cvtColor(img_array,cv2.COLOR_BGR2RGB))
        plt.show()
        break
        
    break
 
```

### change img size

```python
img_size = 224 # imgNet => 224 x 224
new_array = cv2.resize(img_array,(img_size,img_size)) #resizing
plt.imshow(cv2.cvtColor(new_array,cv2.COLOR_BGR2RGB))
plt.show()
```

### Read All the images and convert them to an array

```python
training_data = [] ## data

def create_training_data():
    for category in classes:
        path = os.path.join(Datadirectory,category)
        class_num = classes.index(category)# angry, surprise ...
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img))
                new_array = cv2.resize(img_array,(img_size,img_size))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
```

```python
create_training_data() # call the func
```

```python
print(len(training_data))
```

```python
#temp = np.array(training_data)
```

```python
#temp.shape
```

```python
import random
random.shuffle(training_data)
```

```python
x = [] # data feature eg size
y = [] # labels eg sad

for features,label in training_data:
    x.append(features)
    y.append(label)
    

x = np.array(x).reshape(-1,img_size,img_size,3)# converting it 4 dimensions
```

```python
# normalize the data
x = x/255.0
```

## Deep Learning model for tarining - Transfer Learning

```python
import tensorflow as tf
from tensorflow  import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2() # pre-trained Model
```

```python
#check model summary
model.summary()
```

### Perform Transfer Learning - Tuning, weight will start from late checkpoint

```python
base_input = model.layers[0].input # taking the first layer
```

```python
base_output = model.layers[-2].output # to get the 1280 base layer 
```

```python
base_output
```

```python
final_output = layers.Dense(128)(base_output) #adding new layer after the global pooling layer
final_output = layers.Activation("relu")(final_output) #activation function
final_output = layers.Dense(64)(final_output)
final_output = layers.Activation("relu")(final_output)
final_output = layers.Dense(7, activation="softmax")(final_output)# my classes are 7. its my classification layer.
```

```python
final_output #my output
```

## Create a model

```python
new_model = keras.Model(inputs = base_input, outputs = final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics= ["accuracy"])
```

```python
new_model.fit(x,Y, epochs =5)
```

```python
faces = faceCascade.detectMultiScale(gray,1.1,4)
for x,y,w,h in faces:
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = frame[y:y+h, x:x+w]
    cv2.rectangle(frame,(x,y), (x+w, y+h),(255,0,0), 2)#BGR for blue color
    facess = faceCascade.detectMultiScale(roi_gray)
    if len(facess) == 0:
        print("Face not detected")
    else:
        for(ex,ey,ew,eh) in facess:
            face_roi = roi_color[ey: ey+eh, ex:ex +ew]
```

# REAL TIME VIDEO DEMO

```python
path = "haarcascade_frontalface_default.xml"
font_scale = 1.5
font = cv2.FONT_HERSHEY_PLAIN
rectangle_bgr = (255, 255, 255)
img = np.zeros((500, 500, 3), dtype=np.uint8)
text = "text in a box"
(text_width, text_height), _ = cv2.getTextSize(text, font, fontScale=font_scale, thickness=1)
text_offset_x = 10
text_offset_y = img.shape[0] - 25
box_coords = ((text_offset_x, text_offset_y), (text_offset_x + text_width + 2, text_offset_y - text_height - 2))
cv2.rectangle(img, box_coords[0], box_coords[1], rectangle_bgr, cv2.FILLED)
cv2.putText(img, text, (text_offset_x, text_offset_y), font, fontScale=font_scale, color=(0, 0, 0), thickness=1)

cap = cv2.VideoCapture(0)  # Changed from 1 to 0 to use the default camera (usually the built-in webcam)
if not cap.isOpened():
    raise IOError("Cannot open camera")

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    faces = faceCascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    for x, y, w, h in faces:
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
        facess = faceCascade.detectMultiScale(roi_gray)
        
        if len(facess) == 0:
            print("Face not detected")
        else:
            for (ex, ey, ew, eh) in facess:
                face_roi = roi_color[ey:ey + eh, ex:ex + ew]  # Crop face
    
        final_image = cv2.resize(face_roi, (224, 224))
        final_image = np.expand_dims(final_image, axis=0)
        final_image = final_image / 255.0
    
        Predictions = new_model.predict(final_image)
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 1.5
        
        if np.argmax(Predictions) == 0:
            status = "Angry"
        elif np.argmax(Predictions) == 1:
            status = "Disgust"
        elif np.argmax(Predictions) == 2:
            status = "Fear"
        elif np.argmax(Predictions) == 3:
            status = "Happy"
        elif np.argmax(Predictions) == 4:
            status = "Neutral"
        elif np.argmax(Predictions) == 5:
            status = "Sad"
        else:
            status = "Surprise"
        
        x1, y1, w1, h1 = 0, 0, 175, 75
        cv2.rectangle(frame, (x1, y1), (x1 + w1, y1 + h1), (0, 0, 0), -1)
        cv2.putText(frame, status, (x1 + int(w1 / 10), y1 + int(h1 / 2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
        cv2.putText(frame, status, (100, 150), font, 3, (0, 0, 255), 2, cv2.LINE_4)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)
    
    cv2.imshow("Face Expression Recognition", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
